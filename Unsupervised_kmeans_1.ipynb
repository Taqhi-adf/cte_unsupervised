```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# 1. Generate dataset
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=0.60, random_state=42)

# 2. Compute inertia for different k values
inertia = []
K_range = range(1, 11)

print("K Value vs Inertia (WCSS)")
print("-" * 30)

for k in K_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)  # ✅ Added n_init
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    print(f"k = {k}, Inertia = {kmeans.inertia_:.2f}")

# 3. Plot the elbow method
plt.plot(K_range, inertia, 'bo-')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia (WCSS)")
plt.title("Elbow Method for Optimal k")
plt.show()

# 4. Final clustering with the best k (say k=4)
kmeans = KMeans(n_clusters=4, n_init=10, random_state=42)
y_pred = kmeans.fit_predict(X)

# 5. Plot final clusters
plt.scatter(X[:,0], X[:,1], c=y_pred, cmap="viridis")
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1],
            s=200, c="red", marker="X")
plt.title("KMeans Clustering with k=4")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
# Generate synthetic dataset (if df is unavailable)
df, _ = make_blobs(n_samples=200, centers=4, cluster_std=1.5, random_state=42)
df = pd.DataFrame(df, columns=["Age", "Spending_Score(1-15000)"])
```

```python
df.columns
```

```python
df
```

```python
# Feature selection
X = df[["Age", "Spending_Score(1-15000)"]]
```

```python
# Compute Inertia for Different K Values
inertia = []
k_values = range(1, 11)

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)
    print(f"K={k}, Inertia={kmeans.inertia_:.2f}")
```

```python
# Elbow Method Visualization
plt.figure(figsize=(10, 6))
plt.plot(k_values, inertia, marker="o", linestyle="--", color="red", label="Inertia")
plt.xticks(k_values)
plt.xlabel("Number of Clusters (K)", fontsize=12)
plt.ylabel("Inertia (Sum of Squared Distances)", fontsize=12)
plt.title("Elbow Method: Finding the Optimal K", fontsize=14, fontweight="bold")
```

```python
# Highlight the Elbow Point
optimal_k = 4  # Assume elbow at K=4 for our dataset
plt.scatter(optimal_k, inertia[optimal_k - 1], marker="o", color="blue", s=200, label=f"Optimal K={optimal_k}")
plt.legend()
plt.show()
```

```python
optimal_clusters = 4
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(X)
plt.figure(figsize=(8, 5))
plt.scatter(df["Age"], df["Spending_Score(1-15000)"], c=df["Cluster"], cmap="rainbow", alpha=0.7)
plt.xlabel("Age")
plt.ylabel("Spending Score")
plt.title("K-Means Clustering - Age vs Spending Score")
plt.show()
```
